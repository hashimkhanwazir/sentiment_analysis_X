### It shows that the tweets text is not cleaned. It has hash tags, extra spaces, URLs, and emojis. Let us clean it step by step



```python
### 1. Remove URLs  ###
import re

fireDF["tweet_text"] = fireDF["tweet_text"].apply(lambda x: re.sub(r"http\S+|www\S+|https\S+", '', x, flags=re.MULTILINE))
fireDF
```

```python
### 2. Remove Hashtags ###
fireDF["tweet_text"] = fireDF["tweet_text"].apply(lambda x: re.sub(r"#\S+", '', x))
fireDF
```

```python
# 3. Remove Special Characters and Emojis
fireDF["tweet_text"] = fireDF["tweet_text"].apply(lambda x: re.sub(r"[^a-zA-Z0-9\s]", '', x))
fireDF
```

```python
# 4. Normalize Whitespace
fireDF["tweet_text"] = fireDF["tweet_text"].apply(lambda x: ' '.join(x.split()))
fireDF["tweet_text"]
```

### Since the tweets texts are cleaned, now we will clean the Names as well 

```python
import emoji
```

```python

# Function to remove emojis
# Define the function
def remove_emojis(column, replace=''):
    """
    Remove emojis from a specified column in the DataFrame.
    
    Parameters:
        column (pd.Series): The column from which emojis should be removed.
        replace (str): The string to replace emojis with. Default is an empty string.
    
    Returns:
        pd.Series: The column with emojis removed or replaced.
    """
    return column.apply(lambda x: emoji.replace_emoji(x, replace=replace) if isinstance(x, str) else x)
    # type: ignore
    # Apply the function to the name columns

fireDF['name'] = remove_emojis(fireDF['name'], replace='')
fireDF["name"]
```
